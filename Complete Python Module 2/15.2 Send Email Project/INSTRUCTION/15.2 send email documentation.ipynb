# 15.2 Project: Send Emails using Python



## 15.2.1 Objective: -
* Send emails to the client using python
* Email contains an HTML template
* It also contains a PDF attachment
* It reads the credentials from the config file for reading senders email and passKey

---

## 15.2.2 File Structure:- 

```
.
├── main.py
├── getCredentials
│   ├── config.ini
│   └── read.py
└── sendDetailedEmail
    ├── email.py
    ├── FAQ1.pdf
    └── Template_corona_info.html
    
```

## 15.2.3 Brief Explanation:- 
* **main.py**: This is the main file to be executed 
* **getCredentials**: This module/directory contains two file-
> * **config.ini**: This is a configuration file which contains email id and password of the sender
> * **read.py**: This reads the configuration from from config.ini
* **sendDetailedEmail**: This modile/directory contains 3 files-
> **email.py**:This file attaches the subject line, pdf attachement and HTML template to the email body.
> **FAQ1.pdf**:This contains covid related FAQ. You can replace with any other PDF file.
> **Template_corona_info.html**:This contains HTML template for covid related info. It can be changed as per your requirements
