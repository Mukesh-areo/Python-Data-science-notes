````python
# create a pandas dataframe
# with a dummy data = random 1000 nos generated by numpy.
# with index = data range from date 
df = pd.DataFrame(np.random.randn(1000), 
                  index = pd.date_range('2020-08-08', periods = 1000), 
                  columns=['value'])

# create a series dataframe for time series
# with a dummy data = random 1000 nos generated by numpy.
# with index = data range from date 
ts = pd.Series(np.random.randn(1000), 
               index = pd.date_range('2020-08-08', periods = 1000))
````

#### Let's try to plot IRIS dataset
You can import IRIS dataset from seaborn library which is another visualization library built over matplotlib. We'll see it in detail in Seaborn chapter. But for now let's get the data.

> Iris dataset is a dataset that distinguishes among 3 species of flower based on their sepal length,sepal width,petal length,and petal width. To know about **iris dataset** you can [click here](https://archive.ics.uci.edu/ml/datasets/iris) 

````python
# import seaborn library
import seaborn as sns

# load iris dataset
iris = sns.load_dataset('iris')
iris.head(5)
````

````python
# Lets create normal line plot using plot method on pandas dataframe.
# here we have specified title of the graph.

iris.plot(figsize=(20,10), title = "IRIS Dataset")
````

**NOTE**: As you might have observed it always prints line graph when you use plot method on the dataframes. But if you want to plot other kind of graphs like bar then you can specify another parameter i.e. kind = "bar"

````python
# plotting bar plot for first 10 entry of the iris datset.
# Here we have specified kind = bar to plot the bar graphs.
# Plus we are using iloc method on pandas datframe to get first 10 data points
iris.iloc[:10].plot(figsize=(20,5), title = "IRIS Dataset", kind="bar")
````

````python
# lets try to keep the numerical values only
# use drop method to drop "sepcies" column
iris_num = iris.drop(["species"], axis=1)
````

````python
iris_num.head()
````

````python
# plotting 0th datapoint as a bar
iris_num.iloc[0].plot(kind="bar")
````

````python
# and alternative of above method is you can use bar method 
# instead of using kind parameter
iris_num.iloc[0].plot.bar()
````

### 18.1.1.3 Histograms

it was introduced by *Karl Pearson*. It is a representation of numerical data in a form of bar graphs. Following is a bried explaination: -

* BINS: these are nothing division of entire range of values into some specific series of intervals. These intervals are non overlapping. 
* Then we plot the count of values or frequency of values which fall in these bins or buckets in form of a bar

For more information you can refer - [Histogram (WIKIPEDIA)](https://en.wikipedia.org/wiki/Histogram)

````python
# plot historgram for iris dataset
# this plots the distribution of all the columns in one graph
iris.plot.hist()
````

````python
# you can plot historgam of one column as shown below
iris['sepal_length'].hist()
````

#### you can also specifiy unequally spaced bins by passing a list
in the below example one bin is between 1 and 4 and other one is 
between 4 and 10
* first bin excludes 4 $i.e$ its range $\in\ [1,4)$ 
* second bin or last bin includes both 5 and 10 both $i.e.$ its in range $\in\ [4,10]$

````python
iris['sepal_width'].hist(bins =[1,4,10])
````

````python
# alternative way to plot histogram
iris.plot(kind ="hist")
````

````python
# here we have plotted stacked bar graph with bins = 50
# to observe difference between stacked and unstacked lets plot 
# both the cases
# case 1: stacked = True {default case}
iris.plot(kind = 'hist', stacked = True, bins = 50)
````

````python
# case 2: stacked = False
iris.plot(kind = 'hist', stacked = False, bins = 50)
````

````python
# we can also make stacked horizontal historgrams
# by using orientaion = 'horizontal' parameter
iris.plot(kind = 'hist', stacked = True, bins = 50, orientation = 'horizontal')
````

````python
# Lets calculate the first discrete difference of element
# Its nothing but substracting current cell or row from previous cell
iris['sepal_width'].diff()[:5]
````

````python
# plotting the difference
iris['sepal_width'].diff().plot(kind = 'hist', stacked = True, bins = 50)
````

````python
# drop the species column
iris_df1 = iris.drop(['species'], axis = 1)
````

#### Lets plot the histogram of diff element by- 
* specifying color
* specifying figuresize
* specifying alpha value which is a paramter used for specifying opaqueness of graph
* here, 
> * 0 $\implies$ 100% tansparent
> * 1 $\implies$ 100% opaque

````python
iris_df1.diff().hist(color = 'b', alpha = 0.1, figsize=(10,10))
````
