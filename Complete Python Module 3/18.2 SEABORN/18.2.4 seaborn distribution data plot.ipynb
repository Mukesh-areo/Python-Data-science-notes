### 18.2.4 Visualizing Distribution of the Data
- ```distplot()```
- ```jointplot()```
- ```kdeplot()```
- ```pairplot()```

````python
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
# generate random normal distribution points
x = np.random.randn(100)
````

#### 18.2.4.1 ```distplot()```

Full signature of distplot-

```python
seaborn.distplot(a, bins=None, hist=True, kde=True, rug=False, fit=None, hist_kws=None, kde_kws=None, rug_kws=None, fit_kws=None, color=None, vertical=False, norm_hist=False, axlabel=None, label=None, ax=None)
```

for more information refer here - [diplot](https://seaborn.pydata.org/generated/seaborn.distplot.html)

````python
# lets plot ditribution plot with x
sns.distplot(x, kde = True, hist = False, rug= False, bins= 30)
````

````python
# displor with shade
sns.kdeplot(x, shade=True, cbar = True, bw = 1, cut = 0)
````

````python
# load tips dataset
tips = sns.load_dataset('tips')
tips.head()
````

````python
# get x and y value from tips dataset
x = tips['total_bill']
y = tips['tip']
````

#### 18.2.4.2 ```jointplot()```

Full signature of jointplot-

```python
seaborn.jointplot(x, y, data=None, kind='scatter', stat_func=None, color=None, height=6, ratio=5, space=0.2, dropna=True, xlim=None, ylim=None, joint_kws=None, marginal_kws=None, annot_kws=None, **kwargs)
```
for more information refer here - [jointplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

````python
# lets plot a joint plot which is combo of histogram and scatter plot
sns.jointplot(x = x, y=y)
````

````python
sns.set() # set deafault style
# lets select kind =hex for hexagonal plot
sns.jointplot(x = x, y=y, kind = 'hex')
````

````python
# kind = kde plot
sns.jointplot(x = x, y = y, kind = 'kde')
````

````python
import matplotlib.pyplot as plt
````

#### 18.2.4.3 ```kdeplot()```

Function Signature -

```python
seaborn.kdeplot(data, data2=None, shade=False, vertical=False, kernel='gau', bw='scott', gridsize=100, cut=3, clip=None, legend=True, cumulative=False, shade_lowest=True, cbar=False, cbar_ax=None, cbar_kws=None, ax=None, **kwargs)
```

For more info refer docs here - [kdeplot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html)

````python
f, ax = plt.subplots(figsize = (6,6))
cmap = sns.cubehelix_palette(as_cmap = True, dark = 0, light = 1, reverse= True)
sns.kdeplot(x, y, cmap = cmap, n_levels=60, shade=True)
````

````python
g = sns.jointplot(x, y, kind = 'kde', color = 'm')
g.plot_joint(plt.scatter, c = 'w', s = 30, linewidth = 1, marker = '+')
g.ax_joint.collections[0].set_alpha(0)
````

````python
# load iris datset
iris = sns.load_dataset('iris')
iris.head()
````

#### 18.2.4.4 ```pairplot()```

Function signature

```python
seaborn.pairplot(data, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=True, plot_kws=None, diag_kws=None, grid_kws=None, size=None)
```

For more info refer official docs here - [pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

````python
# pairplot 
sns.pairplot(iris)
````

````python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels = 10)
````
